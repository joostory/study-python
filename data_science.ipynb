### 상관관계

상관관계: 의존성의 정도
```
- x, y의 상관관계

 xx xy
 yx yy

- x, y, z의 상관관계

 xx xy xz
 yx yy yz
 zx zy zz
```

corrcoef는 피어슨 상관계수를 계산

````python
x = [i for i in range(10)]
y = [i**2 for i in range(10)]
z = [100 * np.sin(3.14 * i/100) for i in range(10)]
print(np.corrcoef([x, y, z]))
````

## 차트

````python
import matplotlib.pyplot as plt

years = [i * 10 + 1950 for i in range(7)]
gdp = [67.0, 80.0, 257.0, 1686.0, 6506, 11865.3, 22105.3]
gdp2 = [i * 1000 for i in range(7)]

plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.plot(years, gdp2, color='blue', marker='x', linestyle='solid')
plt.title("GDP per capita")
plt.ylabel("dollars")
plt.show()
````

````python
import matplotlib.pyplot as plt

x = [i for i in range(-10, 10)]
y = [2 * i for i in x]

plt.plot(x, y, color='blue', marker='o', linestyle='solid')
plt.title("y = 2x")
plt.ylabel("y")
plt.xlabel("x")
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

x = [i for i in range(100)]
y = [3 * np.random.rand() for i in x]

plt.plot(x, y, color='red', marker='o', linestyle='solid')
plt.title("y = random")
plt.ylabel("y")
plt.xlabel("x")
plt.axis([-1, 101, -1, 4])
plt.show()
````

````python
import matplotlib.pyplot as plt

x = [x for x in range(-20, 20)]
y1 = [2*t for t in x]
y2 = [t**2 + 5 for t in x]
y3 = [-t**2 - 5 for t in x]

plt.plot(x, y1, 'r--', x, y2, 'g^-', x, y3, 'b.:')
plt.axis([-30, 30, -30, 30])
plt.show()
````

````python
import matplotlib.pyplot as plt

x = [x for x in range(12)]
y = [t**2 for t in x]
z = [t**3 for t in x]

plt.plot(x, x, label='linear')
plt.plot(x, y, label='quadratic')
plt.plot(x, z, label='quibic')
plt.plot(x, [2**t for t in x], label='power')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('My Plot')
plt.legend()
plt.show()
````

````python
import math
import matplotlib.pyplot as plt

x = [t for t in range(360)]
r = [math.radians(t) for t in x]
y1 = [math.sin(t) for t in r]
y2 = [math.cos(t) for t in r]

plt.plot(x, y1, 'b-', x, y2, 'r-')
plt.show()
````
